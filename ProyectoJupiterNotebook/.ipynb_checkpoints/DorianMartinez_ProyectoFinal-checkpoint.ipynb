#### Explicación de los Datos:
La base de datos a utilizar en este proyecto proviene de un pequeño reporte real de la lista de clientes de una empresa dedicada al mantenimiento, instalación, venta y reparación de generadores eléctricos (plantas de luz). El reporte fue creado con la intensión de ser lo suficiente informativo como para poder trabajar con el sin tener que exponer datos críticos que pudieran afectar la privacidad de la empresa.

La empresa cuenta con 4 niveles de personal técnicos, niveles 1 al 3 y MC. Siendo el nivel 1 el mas básico y MC el mas avanzado. También cuenta con niveles de prioridad de cada cliente, yendo del 1 (Life safety) al 8 (Portable/RV/Other).

La meta de este proyecto obtener la información que muestre el panorama actual de la cartera de clientes.

* Cantidad de clientes por cuidad 
* Prioridad dominante de clientes
* Cantidad que requieren su mantenimiento anual, mostrado mes a mes
* Relaciones que pudiera haber entre ciudades y prioridades
* Relaciones que pudiera haber entre ciudades y nivel técnico requerido

La base de datos cuenta con 1818 filas y 8 columnas las cuales se describen a continuación:
1. **CustomerNumber**: Número único de identificación del del cliente.
2. **CustomerServiceZipCode**: Código postal de la dirección donde se encuentra el generador eléctrico.
3. **CustomerServiceCity**: Ciudad donde se encuentra el generador eléctrico.
4. **Priority**: Prioridad que tiene el cliente, va del 1 al 8, siendo el 1. Life Safety el de mayor prioridad.
5. **TechType**: Contiene el tipo de técnico requerido para realizar el mantenimiento.
6. **NextOilChangeDate**: Fecha del próximo mantenimiento.
7. **LastOilChangeDate**: Fecha del último mantenimiento realizado.
8. **CustomerIsActive**: Indica si la cuenta del cliente se encuentra activa o no, si esta inactiva significa que el cliente ya no utiliza nuestros servicios.



**1. Librerias y Dataframe** Cargar las librerías **Pandas**, **Numpy** y **Matplotlib**. Posteriormente, crear un **DataFrame** donde se inserte el archivo `.CSV` de datos, mostrando el contenido del **DataFrame**.

```python
#Importación de librerías
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

#Creacion de DataFrame
clientes_df = pd.read_csv('./Datos/customer_report.csv', low_memory=False)

#Impresión de los Resultados
print("El tamaño total del DataFrame es:",clientes_df.shape)
print("\nLa Información del DataFrame es:\n")
clientes_df.info()
print("\nEl DataFrame es:")
clientes_df.head(10)
```

**2. Indice** Definimos como el **Indice** del **DataFrame** a la columna *`CustomerNumber`* ya que es el la unica columna cuyos valores son unicos, indicando que se haga un ordenamiento a través de la misma

```python
#Definir la Columna "ID Usuario" como el Indice del DataFrame
clientes_df_index =  clientes_df.set_index('CustomerNumber').sort_index()

#Revisando el formato de los datos contenidos en el Indice del DataFrame
print("\nEl DataFrame es:")
clientes_df_index.head(10)
```

**3 Acondicionamiento del Data Frame** 

**3.1 Limpieza**
Eliminar clientes *`inactivos`* o con campos *`TechType`*,  *`Priority`*, *`CustomerServiceZipCode`*  o *`CustomerServiceCity`*  nulos.

```python

#clientes_df_index['CustomerIsActive'].value_counts()
# Reemplazamos los "False" con campos nulos y despues simplemente eliminamos los campos nulos
clientes_df_index['CustomerIsActive'].replace(False, np.nan, inplace=True)
clientes_df_index.dropna(subset=['CustomerIsActive'], inplace=True)
#clientes_df_index['CustomerIsActive'].value_counts()

# Eliminamos los campos nulos de la columan "Priority", "TechType", "CustomerServiceZipCode" y "CustomerServiceCity"
clientes_df_index.dropna(subset=['Priority'], inplace=True)
clientes_df_index.dropna(subset=['TechType'], inplace=True)
clientes_df_index.dropna(subset=['CustomerServiceZipCode'], inplace=True)
clientes_df_index.dropna(subset=['CustomerServiceCity'], inplace=True)

# Comprobamos cambios realizados
print("Contenido de 'False' en CustomerIsActive: ")
print(np.unique(clientes_df_index['CustomerIsActive'], return_counts=True))
print("Cantidad de nulos en columna Priority: ", clientes_df_index['Priority'].isnull().sum())
print("Cantidad de nulos en columna TechType: ", clientes_df_index['TechType'].isnull().sum())
clientes_df_index.shape
```

**3.2 Columna Fecha de Mantenimiento** 

Convertir las columnas *`NextOilChangeDate`* y *`LastOilChangeDate`* a formato de fecha y hora. y combinarlas quedandonos con el dato mas reciente y sustituyendo tantos campos nulos como sea posible. Los campos nulos restantes seran eliminados

```python
#Conversión de las Columnas de "Fechas" al formato de Serie de Tiempo
clientes_df_index['NextOilChangeDate'] = pd.to_datetime(clientes_df_index['NextOilChangeDate'])
clientes_df_index['LastOilChangeDate'] = pd.to_datetime(clientes_df_index['LastOilChangeDate'])

# Rellenamos los espacios vacios de las columnas Last y Next Oil Change date con la fecha contenida en 
# cualquiera de las dos. El resultado final debe de ser que aquella fila que tenia almenos un espacio vacio, 
# ahora tendra o la misma fecha o tendran un NaTs 
clientes_df_index['NextOilChangeDate'] = clientes_df_index['NextOilChangeDate'].fillna(clientes_df_index['LastOilChangeDate'])
clientes_df_index['LastOilChangeDate'] = clientes_df_index['LastOilChangeDate'].fillna(clientes_df_index['NextOilChangeDate'])

# De las fechas de las dos columnas nos quedamos con la ultima y la sustituimos en 'NextOilChangeDate'
clientes_df_index.sort_values('NextOilChangeDate').drop_duplicates('LastOilChangeDate', keep='last')

# Eliminamos los espacios vacios
clientes_df_index.dropna(subset=['NextOilChangeDate'], inplace=True)

# Ya no necitaremos la columna "LastOilChangeDate" asi que la eliminamos
del clientes_df_index['LastOilChangeDate']

# El nombre de la columna "NextOilChangeDate" no es preciso, asi que lo reemplazamos con "MaintenanceDate"
clientes_df_index.rename(columns = {'NextOilChangeDate': 'MaintenanceDate'}, inplace = True)

# Imprimimos resultados
clientes_df_index.head(12)
```

**3.3 Columna TechType** 

El dropdown menu de la aplicacion que llena la columna de "TechType" fue actualizado en algun punto despues de que ya habia datos, es por ello que algunas filas muestran solo el nivel del tecnico y otras  muestran el nivel y un ejemplo en parentesis, para nuestros fines practicos nos combiene equializar la informacion y eliminar los campos vacios

```python
#Realizamos un mapeo de los valores de la columna "TechType" 
clientes_df_index['TechType'] = clientes_df_index['TechType'].map({
        'Level 1 (i.e. air cooled)':'Level 1',
        'Level 1':'Level 1',
        'Level 2 (i.e. liquid cooled)':'Level 2',
        'Level 2':'Level 2',
        'Level 3 (i.e. diesel)':'Level 3',
        'Level 3':'Level 3'},
         na_action= 'ignore')

clientes_df_index.tail(10)
```

En el paso anterior se finalizo con el proceso de acondicionamiento del DataFrame, ahora solo resta analizarlo y presentar conclusiones

En esta grafica se puede observar que la mayor cantidad de clientes se encuentran en San Jose, la cuidad mas industrializada del

**4.2 Cantidad de clientes por tipo de tecnico**

En la practica, los tecnicos Nivel 2 son los mas dinamicos ya que pueden atender a la mayoria de clientes comerciales e industriales.
```python
# Configuramos el grafico
plt.figure(figsize=(8,6))
plt.title("Cantidad de clientes por tipo de tecnico")
plt.ylabel("Nivel de tecnico requerido")
plt.xlabel("Numbero de clientes")
# Graficamos los datos
clientes_df_index['TechType'].value_counts().plot(kind = 'barh')
plt.savefig('./Graficos/NivelTecnicoRequerido.png')
plt.show()
```

**4.3 Cantidad que requieren su mantenimiento anual, mostrado mes a mes**

Esta grafica muestra la distribución de mantenimientos a lo largo del año. La grafica muestra información que coincide con lo que vemos en la realidad. Excepto por los meses de Octubre y Noviembre los cuales se utilizan para darle mantenimiento a unos pocos clientes que tienen una enorme cantidad de generadores.

```python
import calendar
# Se utilizara este arreglo como index del grafico
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
#Copiamos el Data Frame
MesMant_df = clientes_df_index
# Obtenemos solo el mes de la fecha en la columan "MaintenanceDate"
MesMant_df['Month'] = MesMant_df['MaintenanceDate'].dt.month

#Graficamos
temp = MesMant_df['Month'].astype(int).apply(lambda x: calendar.month_abbr[x]).value_counts().reset_index()
temp.set_index('index', inplace=True)
temp = temp.reindex(index=months)
temp.plot(kind='bar', figsize=(8,8), title="Cantidad de Mantenimientos por Mes")
plt.savefig('./Graficos/MantenimientosPorMes.png')
```

**4.4 Relacion entre las Ciudades, Prioridades, Nivel Tecnico y Mes de Servicio**

Hice esto en varios bloques por que algunas operaciones requerian reiniciar el kernel. La conclusion se muestra un bloque antes de la grafica.

```python
#Importamos libreria para crear la grafica de correlacion
import seaborn as sns

Relacion_df = clientes_df_index

#Eliminamos las columnas que no se utilizaran
del Relacion_df['CustomerServiceCity']
del Relacion_df['CustomerIsActive']
del Relacion_df['MaintenanceDate'] 
#Relacion_df.info()
```

```python
# Limpiamos la cadena para obtener solo el dato numerico de las columnas "TechType" y "Priority"
Relacion_df['TechType'] = Relacion_df['TechType'].str.extract(r'(\d+[.\d]*)')
Relacion_df['Priority'] = Relacion_df['Priority'].str.extract(r'(\d*)')

#Convertimos los valores a tipo "int" para poderlos graficar
Relacion_df['TechType'] = Relacion_df['TechType'].astype(int)
Relacion_df['Priority'] = Relacion_df['Priority'].astype(int)
Relacion_df.head(10)
```

```python
# Algunos codigos postales cuentan con el nuevo formato de mas de 5 digitos (ej. 94541-3922) para fines practicos se
# converiran a modo clasico y posteriormente a un tipo de variable entero

#Esta funcion extrae los primeros 5 caracteres de la cadena
def recortar_zipcode(zip):
      return zip.astype(str).str.extract('(\d+)', expand=False).str.zfill(5)

#Obtenemos los primeros 5 caraceres de la cadena 
Relacion_df['CustomerServiceZipCode'] = recortar_zipcode(Relacion_df['CustomerServiceZipCode'])
# Convertimos las cadenas a tipo "int"
Relacion_df['CustomerServiceZipCode'] = Relacion_df['CustomerServiceZipCode'].astype(int)

# Renombramos la columna para que se aprecie mejor en el grafico
clientes_df_index.rename(columns = {'CustomerServiceZipCode': 'Zip Code'}, inplace = True)


#Comprobamos que todas las variables son del tipo "int"
Relacion_df.info()
```

**4.4.1 Relacion entre las Ciudades, Prioridades, Nivel Tecnico y Mes de Servicio (conclusion)**

Me sorprendio ver la alta relacion entre el tipo de tecnico y el mes, despues de pensarlo un poco me doy cuenta de que si es correcto, los meses de Oct, Nov y Dec los dedicamos a cuentas Industriales y Live Safety, las cuales requieren tecnicos de niveles 2 y 3, los meses de Abril y Marzo por lo regular es cuando los clientes Residenciales prefieren hacer sus manteniemientos. Me parecio muy interesante esto de la correlacion. 
Originalmente esperaba que ubiera una mayor relacion entre el tipo de tecnico y el zip code, esto por que hay zonas residenciales (Tecnicos Nv.1) y zonas industriales/comerciones (Tecnicos Nv.2 y 3). Que al final de cuentas es la segunda correlacion con mas peso, pero es solo un 58% comparado con el 76% del mes/tecnico/

```python
# Finalmente graficamos los datos

#Generación de la Matriz de Correlación
mat_corr = sns.heatmap(Relacion_df.corr(), annot = True, cmap = 'cool')
plt.title('Relacion entre Cuidades, Nivel Tec y Fecha de Mant.')
plt.show()

#Guardamos el grafico en un png
grafico = mat_corr.get_figure()  
grafico.savefig('./Graficos/Correlacion.png')
```
